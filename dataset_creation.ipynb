```python
clean_dir = "clean/"
clean_files = listdir(clean_dir)
np.random.shuffle(clean_files)
noise_dir = "noise/"
noise_files = listdir(noise_dir)
np.random.shuffle(noise_files)
mixed_dir = "mixed/"
print(noise_files)
```

```python
def signal_mixing(clean_fn, noise_fn, clean_dir, noise_dir, output_dir):
    # Define SNR range and choose a random SNR for mixing
    min_snr = 10  # Minimum SNR in dB
    max_snr = 20  # Maximum SNR in dB
    random_snr = np.random.uniform(min_snr, max_snr)
    # Calculate the scaling factor for the noise based on SNR
    scaling_factor = min_snr ** (-random_snr / float(max_snr))
    clean_audio, sr1 = librosa.load(f"{clean_dir}{clean_fn}", duration=5.)
    noise_audio, _ = librosa.load(f"{noise_dir}{noise_fn}", duration=5.)
    # Mix the clean bird song and noise
    mixed_audio = clean_audio + scaling_factor * noise_audio
    # Normalize the mixed audio
    mixed_audio /= np.max(np.abs(mixed_audio))

    sf.write(f"{output_dir}{clean_fn}", mixed_audio, sr1)
```

```python
for i, noise in enumerate(noise_files):
    signal_mixing(clean_files[i], noise, clean_dir, noise_dir, mixed_dir)
    
```
